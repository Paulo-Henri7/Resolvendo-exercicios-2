#1. Seleção de Dados

1.1. Carregue o arquivo results.csv em um DataFrame e exiba as 10 primeiras linhas.

```python
import pandas as pd
```

```python
url_results = "/content/results.csv"
resultados = pd.read_csv(url_results)
resultados.head(10)
```

1.2.Selecione todas as partidas que ocorreram no Brasil.

```python
resultados[resultados["country"] == "Brazil"]
```

1.3. Filtre apenas as partidas onde o time da casa marcou mais de 3 gols.

```python
resultados[resultados["home_score"] > 3][["home_team","home_score"]]
```

#2.Agrupamento e Agregação

2.1. No arquivo goalscorers.csv, conte quantos gols cada jogador marcou no total.

```python
gols = pd.read_csv("/content/goalscorers.csv")
gols_por_jogador = gols.groupby("scorer").size().reset_index(name="total_gols")
gols_por_jogador.sort_values(by="total_gols", ascending=False)
```

2.1. No arquivo results.csv, calcule a média de gols marcados pelos times da casa agrupando por torneio.

```python
resultados.groupby(["tournament"])[["home_score"]].mean().sort_values(by="home_score", ascending=False)
```

#3. Junção de Dados

3.1.Use os arquivos results.csv e goalscorers.csv para encontrar quais partidas tiveram gols e liste os times e os artilheiros.

```python
partidas = resultados.merge(gols, on=["date", "home_team", "away_team"])
partidas[["home_team","away_team","home_score","away_score","scorer"]]
```

3.2.Combine os dados de former_names.csv com results.csv para identificar partidas de times que mudaram de nome.

```python
nomes_times = pd.read_csv("/content/former_names.csv")
nomes_antigos = resultados.merge(nomes_times, left_on="home_team", right_on="current")
nomes_antigos[nomes_antigos["former"].notna()][["date","home_team","former"]]
```

#4. Transformações e Estatísticas

4.1. No arquivo shootouts.csv, descubra qual seleção venceu mais disputas de pênaltis.

```python
selecao_vencedora = pd.read_csv("/content/shootouts.csv")
vitoria = selecao_vencedora["winner"].value_counts().sort_values(ascending=False).reset_index(name="vitorias")
vitoria.head(50)
```

4.2.o arquivo results.csv, calcule a porcentagem de jogos que terminaram empatados.

```python
empates= pd.read_csv("/content/results.csv")
empates
```

```python
total_partidas = len(empates)
empates_totais = len(empates[empates["home_score"] == empates["away_score"]])
porcentagem_empates = (empates_totais/total_partidas)*100
print(f"O total de partidas é: {total_partidas}. E o total de partidas empatadas é: {porcentagem_empates:.2f}%")
```
